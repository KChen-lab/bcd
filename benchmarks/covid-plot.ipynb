```python
import scanpy as sc
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import scib_metrics
from scib_metrics.benchmark import Benchmarker
```

```python
methods = ['bcd_umap', 'pca_umap', 'limma_umap', 'seurat_umap', 'seurat_rpca_umap', 'harmony_umap']
sample = 'gse145926'

metadata = pd.read_csv(f"../output-embeddings/gse145926_label.csv", index_col=0)
```

```python
embs = {}
for m in methods:
    embs[m] = pd.read_csv(f"../output-embeddings/{sample}_{m}.csv", index_col=0)
```

```python
adata = sc.AnnData(np.zeros([len(metadata), 0]), obs=metadata)
adata
```

```python
for m in methods:
    matrix_name = m
    adata.obsm[matrix_name] = embs[m].to_numpy()
```

```python
adata.obs
```

```python
fig, axes = plt.subplots(4, 4, figsize=(9, 8))
for i, m in enumerate(methods):
    legend_loc = None
    if i == 1:
        legend_loc = 'right margin'
    j = i
    k = 0
    if j >= 2:
        k = 1
        j = j - 2
    sc.pl.embedding(adata, m, color='sample_new', ax=axes[k * 2][j], legend_loc=legend_loc, title='', legend_fontsize=10, 
                    alpha=0.75, size=1.5)
    sc.pl.embedding(adata, m, color='celltype', ax=axes[k * 2 + 1][j], legend_loc=legend_loc, title='', legend_fontsize=10, 
                    alpha=0.75, size=1.5)
    if i == 1:
        axes[(i // 4) * 2][i % 4].legend(ncol=2, bbox_to_anchor=(1., 1.), frameon=False, fontsize=10)
        axes[(i // 4) * 2 + 1][i % 4].legend(ncol=3, bbox_to_anchor=(1., 1.), frameon=False, fontsize=10)
        
for i in range(2, 4):
    axes[(i // 4) * 2][i % 4].set_axis_off()
    axes[(i // 4) * 2 + 1][i % 4].set_axis_off()
# plt.tight_layout(pad=1.5)
fig
```
