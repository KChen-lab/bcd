```python
metadata = pd.read_csv(f"../output-embeddings/sim_eq_metadata.csv", index_col=0)
nonoise = pd.read_csv(f"../output-embeddings/sim_eq_pca_nonoise.csv", index_col=0)
# pca = pd.read_csv(f"../output-embeddings/sim_eq_pca.csv", index_col=0)
# bcd = pd.read_csv(f"../output-embeddings/sim_eq_bcd.csv", index_col=0)
```

```python
metadata['label'].unique()
```

```python
metadata['sample_label'] = metadata['label'] + '_' + metadata['sample']
metadata
```

```python
categories_A = [
              'S->A_day1', 
              'S->A_day2_sample1', 'S->A_day2_sample2', 
              'A_day2_sample1',  'A_day2_sample2',
              'A_day3_sample1',  'A_day3_sample2',
              'A->A1_day3_sample1', 'A->A1_day3_sample2',
              'A->A2_day3_sample1', 'A->A2_day3_sample2',
              'A1_day4_sample1', 'A1_day4_sample2',
              'A2_day4_sample1', 'A2_day4_sample2',
              ]
categories = list(reversed(categories_A)) + ['S_day1'] + [i.replace('A', 'B') for i in categories_A]
len(categories)
```

```python
categories
```

```python
metadata['order'] = metadata.sample_label.apply({v: i for i, v in enumerate(categories)}.__getitem__)
```

```python
keep_idx = []
for i in metadata['sample_label'].unique():
    keep_idx.extend(metadata.index[metadata['sample_label'] == i][:5])
```

```python
metadata = metadata.sort_values(by='order')
```

```python
metadata = metadata[metadata.index.isin(keep_idx)]
nonoise = nonoise.loc[metadata.index].iloc[:, :5]
nonoise
```

```python
import sklearn as skl
nonoise_dist = skl.metrics.pairwise_distances(nonoise)
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
sns.heatmap(np.log1p(nonoise_dist), vmin=0, vmax=3, ax=ax, cbar=False, xticklabels=False, yticklabels=False)
for i in range(0, 161, 5):
    plt.axhline(i, c='white', lw=1.)
    plt.axvline(i, c='white', lw=1.)
    
def shorten(s):
    return s.replace('day', 'd').replace('sample', 's')
short_categories = list(map(shorten, categories))

plt.xticks(np.arange(2.5, 155, 5), short_categories, rotation=90, size=9, ha='center')
plt.yticks(np.arange(2.5, 155, 5), short_categories, rotation=0, size=9, va='center')
pass
```
